```python
import cv2
import numpy as np
import matplotlib.pyplot as plot
import pkg_resources
%matplotlib inline
```

```python
pic = cv2.imread('../serverGo/output/face.jpg', 0)
cairo = cv2.imread('../serverGo/output/Cairo_conference.jpg', 0)
```

```python
plot.imshow(cairo, cmap='gray')
```

```python
def face_detect(img):
    haar_xml = pkg_resources.resource_filename('cv2', 'data/haarcascade_frontalface_default.xml')
    front_face_cas = cv2.CascadeClassifier(haar_xml)
    face_img = img.copy()
    rect = front_face_cas.detectMultiScale(face_img)
    for (x,y,w,h) in rect:
        cv2.rectangle(face_img, (x,y), (x+w, y+h), (255,255,255), 6)
    return face_img
```

```python
result = face_detect(cairo)
```

```python
plot.imshow(result, cmap='gray')
```
